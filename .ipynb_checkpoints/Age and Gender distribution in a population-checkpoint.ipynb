```python
# Gender's distribution in a population 
categories = ['Male', 'Female', 'Others']
counts = [220, 250, 45]
plt.bar(categories, counts, color=['red', 'blue', 'green'])
plt.grid(axis='y', linestyle='--', alpha=0.7) # # adding horizontal grid lines 
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("Gender's Distribution in a Population")
plt.show()
```

```python
# Age Distribution in a population
ages = np.random.normal(40, 15, 200)  # Generate random ages with mean=40, std deviation=15 and size=200
ages = np.clip(ages, 0, 100) # fixing ages between 0 to 100
plt.hist(ages, bins=10, color='green', edgecolor='black', alpha=0.7) 
plt.grid(axis='y', linestyle='--', alpha=0.7) # adding horizontal grid lines 
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in a population')
plt.show()

     
```
